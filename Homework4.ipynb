```python
plt.figure(figsize=(10, 6))
data_df['What_is_your_major?'].value_counts().plot(kind='bar', color='green')
plt.title('Major Count')
plt.xlabel('Major')
plt.ylabel('Number of:')
plt.show()
# This visual shows us that the most common major in the class in not shockingly Computer Science.
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x='What_is_your_favorite_type_of_food?', y='What_state/country_were_you_born_in?', data=data_df)
plt.title('Favorite Food by Birthplace')
plt.xlabel('Favorite Food')
plt.ylabel('Place of Birth')
plt.show()
# This plot demonstrates the favorite foods depending on the birth place of students
```
