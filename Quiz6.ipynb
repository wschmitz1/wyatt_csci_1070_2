```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
import seaborn as sns
import seaborn.matrix as smatrix
```

```python
data_df = pd.read_csv("titanic.csv")
data_df.head()
data_df
```

```python
data_df.columns
data_df.dtypes
data_df.info()
```

```python
data_df['Fare'] = data_df['Fare'].astype(int)
```

```python
cols = ['SibSp','Parch','Ticket']
data_df = data_df.drop(cols, axis=1)
```

```python
def replace_Sex(val):
    if val == "female":
        return 0
    else:
        return 1
    
data_df['Sex'] = data_df['Sex'].apply(replace_Sex, 1) 
```

```python
data_df['Embarked'].mask(data_df['Embarked'] == 'C', 'Cherbourg', inplace=True)
data_df['Embarked'].mask(data_df['Embarked'] == 'S', 'Southhampton', inplace=True)
data_df['Embarked'].mask(data_df['Embarked'] == 'Q', 'Queenstown', inplace=True)

data_df.head()
```

```python
correlation = data_df.corr()
sns.heatmap(correlation, annot=True)
```

```python
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data_df['Fare'].fillna(data_df['Fare'].mean(), inplace=True)
data_df['Pclass'].fillna(data_df['Pclass'].mean(), inplace=True)
data_df['Age'].fillna(data_df['Age'].mean(), inplace=True)
```

```python
X = data_df[['Fare','Pclass','Age']]
y = data_df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
regression = LogisticRegression()
regression.fit(X_train, y_train)
```

```python
y_pred = regression.predict(X_test)
y_pred
```

```python
accuracy_score = regression.score(X_test, y_test)
accuracy_score
```
